```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv("test_scores.csv")
df.head(3)
```

```python
x = np.array(df.math)
y = np.array(df.cs)
x, y
```

```python
def gradient_descent(x, y):
    m_curr = b_curr = 0
    learning_rate = 0.0002
    iteration = 1
    n = len(x)
    prev_cost = 0
    while(1):
        y_predicted = m_curr * x + b_curr
        cost = (1 / n) * sum([val ** 2 for val in (y - y_predicted)])
        md = -(2 / n) * sum(x * (y - y_predicted))
        bd = -(2 / n) * sum(y - y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        if(math.isclose(cost, prev_cost, rel_tol=1e-20)):
            break
        prev_cost = cost
        if iteration % 1000 == 0:
            print(f"Iteration {iteration}, m: {m_curr:.4f}, b: {b_curr:.4f}, cost: {cost:.6f}")
        iteration += 1
    return m_curr, b_curr, iteration
```

```python
m, b, itr = gradient_descent(x, y)
print(f"m: {m}, b: {b}, iteration: {itr}")
```
